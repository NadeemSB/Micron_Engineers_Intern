```
import docx2txt
gate_pass = list()
date = list()
list1 = list()
ran = range(1300,1802)
for j in ran:
  list1.append(str(j)+".docx")
print(list1)
for i in list1:
  my_text = docx2txt.process(i)
  #MGP
  x = my_text.find("MGP")
  d = my_text.find("DA")
  if (x != -1):
    y = my_text [x:x+100]
    c = y.replace("\n"," ")
    gate_pass.append(c[7:18])
    a = y.find("D")
    b = y[a+3:a+13]
    date.append(b)
  #DA no
  elif (d != -1) :
    d = my_text.find("DA")
    e = my_text[d:d+100]
    f = e.replace("\n"," ")
    gate_pass.append(f[6:16])
    g = e.find("Dt")
    if (g != -1):
      h = e[g+3:g+14]
      date.append(h)
    else :
      k = e.find("DT")
      l = e[k+3:k+13]
      date.append(l)
    #LTCC AND DRC
  else:
    gate_pass.append(" ")
    date.append(" ")
print(gate_pass)
print(date)
```

```
import pandas as pd
df1 = pd.DataFrame()
df1['GATE PASS'] = gate_pass
df1['DATE'] = date
df1.head(30)
```

```
df1.to_excel("Gate_Pass_no.xlsx")
```
